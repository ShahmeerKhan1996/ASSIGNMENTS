```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('factbooks.db')
conn
```

```python
def read_sql(q):
    return pd.read_sql_query(q,conn)
```

```python
q  = '''select * from sqlite_master '''
read_sql(q)
```

### 1- List all cities where the area is greater than 5000, along with their population.

```python
q='''SELECT c.name, c.population, f.area
FROM cities c
INNER JOIN facts f ON c.facts_id = f.id
WHERE f.area > 5000'''

read_sql(q)
```

### 2- Get the name and population of all cities where the population growth rate is greater than 3%.

```python
q='''SELECT c.name, c.population, f.population_growth
FROM cities c
INNER JOIN facts f ON c.facts_id = f.id
WHERE f.population_growth > 3'''

read_sql(q)
```

### 3- Write a query to find the top 5 average population (rounded to 2 decimals) of cities with a population growth rate between 1% and 5%. Group the result by population_growth.

```python
q='''SELECT f.population_growth, round(AVG(c.population),2) AS avg_population
FROM cities c
INNER JOIN facts f ON c.facts_id = f.id
WHERE f.population_growth BETWEEN 1 AND 5
GROUP BY f.population_growth
order by avg_population desc limit 5


'''

read_sql(q)
```

### 4- Find countries that start with "A" and do not have any cities listed in the cities table. The query should return the country name and city name.

```python
q='''SELECT f.name AS country_name, c.name AS city_name
FROM facts f
LEFT JOIN cities c ON f.id = c.facts_id
WHERE f.name LIKE 'A%' AND c.id IS NULL'''

read_sql(q)
```

### 5- List all capital cities and their corresponding death rate, including cities without a matching country. (show in desc order)

```python
q='''SELECT c.name AS city_name, f.death_rate
FROM cities c
LEFT JOIN facts f ON c.facts_id = f.id
WHERE c.capital = 1
ORDER BY f.death_rate DESC
'''

read_sql(q)
```
