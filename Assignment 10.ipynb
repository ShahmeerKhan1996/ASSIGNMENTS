```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('factbooks.db')
conn
```

```python
def read_sql(q):
    return pd.read_sql_query(q,conn)
```

```python
q  = '''select * from sqlite_master '''
read_sql(q)
```

### 1- List all cities where the area is greater than 5000, along with their population.

```python
q='''SELECT c.name, c.population, f.area
FROM cities c
INNER JOIN facts f ON c.facts_id = f.id
WHERE f.area > 5000'''

read_sql(q)
```

### 2- List the names of capital cities where the death rate is greater than 2% and the birth rate is less than 20%

```python
q='''SELECT c.name AS city_name, f.birth_rate, f.death_rate
FROM cities c
INNER JOIN facts f ON c.facts_id = f.id
WHERE c.capital = 1 
  AND f.death_rate > 2 
  AND f.birth_rate < 20'''

read_sql(q)
```

### 3- "List the cities with a population growth rate between 1% and 5%, sorted in desc by population growth."

```python
q='''SELECT c.name AS city_name, c.population, f.population_growth
FROM cities c
INNER JOIN facts f ON c.facts_id = f.id
WHERE f.population_growth BETWEEN 1 AND 5
ORDER BY population_growth
desc
'''

read_sql(q)
```

### 4- Find countries that start with "A" and do not have any cities listed in the cities table. The query should return the country name and city name. Using left join

```python
q='''SELECT f.name AS country_name, c.name AS city_name
FROM facts f
LEFT JOIN cities c ON f.id = c.facts_id
WHERE f.name LIKE 'A%' AND c.id IS NULL'''

read_sql(q)
```

### 5- "List all capital cities and their corresponding migration rate using a LEFT JOIN."

```python
q='''SELECT c.name AS city_name, f.death_rate
FROM cities c
LEFT JOIN facts f ON c.facts_id = f.id
WHERE c.capital = 1
ORDER BY f.death_rate DESC
'''

read_sql(q)
```
